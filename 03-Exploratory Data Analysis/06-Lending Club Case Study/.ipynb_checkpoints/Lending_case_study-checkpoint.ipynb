## <font color="red"> **2.5. Functions** </font>
**Python functions to Clean and Manipulate the data set **

## * strip_date(date,value)
Function to extract Year, month and date

```python
def strip_date(date,value):
    time_stamp = pd.to_datetime(datetime.datetime.strptime(date,'%b-%y'))
    year = time_stamp.year
    month = time_stamp.month
    day = time_stamp.day
    if value == 'year':
        return year
    if value == 'month':
        return month
    return day
    
```

## * clean_and_manipule(df)
Data Cleaning and Manipulation

```python
def clean_and_manipule():
    #Read Dataset
    df = pd.read_csv('./loan.csv')
    
    #Drop Columns:
    null_columns = df.columns[df.isnull().sum() == 39717].tolist()
    Not_useful_columns = ['application_type','initial_list_status','pymnt_plan', 'url', 'desc', 'title', 'zip_code', 'addr_state', 'next_pymnt_d']
    
    df = df.drop(columns=null_columns)
    
    ## Drop higher percentage of NULL value columns
    df = df.drop(columns=['collections_12_mths_ex_med'])
    df = df.drop(columns=['chargeoff_within_12_mths'])
    df = df.drop(columns=['tax_liens'])
    df = df.drop(columns=['mths_since_last_record'])
    df = df.drop(columns=['mths_since_last_delinq'])
    
    ## Drop 0 or 1 Value columns
    columns_with_0_or_1 =['policy_code','acc_now_delinq','delinq_amnt','out_prncp','out_prncp_inv']
    df = df.drop(columns=columns_with_0_or_1)
    
    ## Drop Not useful columns
    df = df.drop(columns=Not_useful_columns)
    df = df.drop(['id'], axis=1)
    
    ###Impute NULL column values:
    df.emp_length = df.emp_length.fillna('< 1 year')
    df.last_credit_pull_d = df.last_credit_pull_d.fillna(df.last_credit_pull_d.mode()[0])
    df.last_pymnt_d = df.last_pymnt_d.fillna(df.last_pymnt_d.mode()[0])
    df.revol_util = df.revol_util.fillna(df.revol_util.mode()[0])    
    
    #pub_rec_bankruptcies
    df.pub_rec_bankruptcies = df.pub_rec_bankruptcies.fillna(df.pub_rec_bankruptcies.mode()[0])
    df.pub_rec_bankruptcies = np.int16(df.pub_rec_bankruptcies)
    
    #Manipulate Columns
    df.int_rate = df.int_rate.apply(lambda x: round(float(x[0:-1]),0)) ##int_rate - Removing the decimal places for simplification of analysis
    df.revol_util = df.revol_util.apply(lambda x: round(float(x[0:-1]),0)) ##revol_util -  Removing the decimal places for simplification of analysis

    #emp_length
    df.emp_length = df.emp_length.replace('< 1 year',0)
    df.emp_length = df.emp_length.apply(lambda x : int(str(x)[0:2].strip()))
    
    #DATE Format
    df['issue_d_year'] = df.issue_d.apply(lambda x: strip_date(x,'year'))
    df['issue_d_month'] = df.issue_d.apply(lambda x: strip_date(x,'month'))
    
    df['earliest_cr_line_year'] = df.earliest_cr_line.apply(lambda x: strip_date(x,'year'))
    df['earliest_cr_line_month'] = df.earliest_cr_line.apply(lambda x: strip_date(x,'month'))
    
    df['last_credit_pull_d_year'] = df.last_credit_pull_d.apply(lambda x: strip_date(x,'year'))
    df['last_credit_pull_d_month'] = df.last_credit_pull_d.apply(lambda x: strip_date(x,'month'))
    
    df['last_pymnt_d_year'] = df.last_pymnt_d.apply(lambda x: strip_date(x,'year'))
    df['last_pymnt_d_month'] = df.last_pymnt_d.apply(lambda x: strip_date(x,'month'))
    
    ##Drop Original Date Columns
    original_date_columns = ['issue_d', 'earliest_cr_line', 'last_credit_pull_d', 'last_pymnt_d']
    df = df.drop(columns=original_date_columns)
        
    # Change
    df.funded_amnt_inv = df.funded_amnt_inv.astype(int)
    df.annual_inc = df.annual_inc.astype(int)
    df.installment = df.installment.astype(int)
    
    # Data Analysis Part changes 
    df = df[~(df.loan_status == 'Current')]
    # For simplify the Analysis changing the Fully Paid = 1 and Charged off = 0
    df.loan_status = df.loan_status.replace('Fully Paid', 1)
    df.loan_status = df.loan_status.replace('Charged Off', 0) # Assumption : Considering Charged Off as Default
    
    df.home_ownership = df.home_ownership.replace('NONE','RENT')
    df.home_ownership = df.home_ownership.replace('OTHER','RENT')
    
    #Derived Column
    df['loan_amnt_5k'] = round(df.loan_amnt / 5000) *5000
    df['annual_inc_10k'] = round(df.annual_inc / 10000) *10000
    
    df['dtiq'] = pd.qcut(df.dti, q = 6, labels=[5,10,15,20,25,30])
    df.dtiq = df.dtiq.astype(int)
    
    ##Since the Current Status Applicant can't be considered either as Defulted or Fully Paid, as such these rows will be ignored
    df = df[~(df.loan_status == 'Current')]
    
    return df
    
```

```python
df = clean_and_manipule()
```

## * Segment Univariate Analysis Final Dataset

```python
def create_segment_df(df_source, column_name, name):
    if column_name =='annual_inc_10k':
        df_pivot = df_source[(df_source.annual_inc_10k<=100000) & (df_source.annual_inc_10k>=10000)].pivot_table(index=['annual_inc_10k'],columns='loan_status',values ='member_id', aggfunc='count')
    else:
        df_pivot = df_source.pivot_table(index=column_name,columns='loan_status',values ='member_id', aggfunc='count')
    
    df = pd.DataFrame(df_pivot[0]/(df_pivot[0]+df_pivot[1])*100)
    df.reset_index(level=0,inplace=True)
    df_segment = pd.DataFrame()
    df_segment['name'] = name
    df_segment['category'] = df[column_name]
    df_segment['result'] = round(df[0])
    df_segment['column_name'] = column_name
    df_segment.fillna(value=name,inplace=True)
    return df_segment
```

**The Loan_status column gives us the clear indication that if the borrower paid the loan or defaulted. As described in the below table, the <font color='red'>Chareged off / Defaulted</font> status indicated that the applicate is defaultd. So we need to analyse the parameter associate with these borrowes and understand correlation**

Since the **Current** Status Applicant can't be considered either as Defulted or Fully Paid, as such these rows will be ignored 

Loan_Status | Description | No of Rows
:-|:-|-:
**Fully paid** |Applicant has fully paid the loan (the principal and the interest rate) | 32950
**Current** | Applicant is in the process of paying the instalments, i.e. the tenure of the loan is not yet completed. These candidates are not labelled as 'defaulted' | 1140
**Charged Off / Defaulted** | Applicant has not paid the instalments in due time for a long period of time, i.e. he/she has defaulted on the loan | 5627

dtiq - Derived column | 30 | 25 | 20 | People with more dti having more chance getting defaulted

# <a id='4'>4. Conclusion and Recommendation to Investors</a>

### <font color='red'>**Multivariate Analysis Top 3 conclution **</font>
- **Upto 38%** Applicants with **lower Grade(lower credit history)**, and if given an **higer loan amount** having high probability of getting defaulted
- **Upto 36%** applicants with **mid tier income**, and if given loan at **above 15% interest rate**, having high probability of getting defaulted
- **Upto 33%** applicants with **higher Debt to Income**, and if loan takeng for the purpose of **Small Business, having high probability of getting defaulted

### <font color='red'>** Recommendation to Inverstors **</font>

- Applicatant with Lower Credit Grade, should not be given higher loan amount
- Applicatant with Mid Tier income(30000 to 60000) should be charged higher interest rate and reduce the loan amount
- Applicatant with Higher Debt to Income should not be give loan for Small Business purpose, and reduce the loan amount
- Applicatant should not be charged more than 19% reduce the probability of becoming defaulter
