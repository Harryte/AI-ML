```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./odi-batting.csv')
```

```python
df.shape
```

```python
df.head(10)
```

```python
df['Century'] = df.Runs >= 100
```

```python
df.groupby(by='Player')['Century'].value_counts()
```

```python
df_pvt = df.pivot_table(index='Player', values='Century', aggfunc='sum')
```

```python
df_pvt.sort_values('Century', ascending= False)
```

```python
df['Strike_rate'] = df.Runs/df.Balls*100
```

```python
df_century = df[(df.Century == True)]
```

```python
df_century.sort_values('Strike_rate', ascending=False)
```

```python
df.MatchDate = pd.to_datetime(df.MatchDate)
```

```python
df['year'] = df['MatchDate'].dt.year
```

```python
df_grp_country_year = df.groupby(by=['Country','year'])
```

```python
df_grp_country_year.Century.value_counts()
```
