## 1.4. Generic Functions

### * roundup_to
Round Up of the value to Nearest Number

```python
def roundup_to(value, round_to_num):
    return int(math.ceil(value / round_to_num)) * round_to_num
```

### * Preprocessing
All the preprocessing steps are included in the function to easy to run this code everytime we open this notebook

```python
def preprocessing(df_bike = np.NaN):
    if(df_bike != np.NaN):
        df_bike = pd.read_csv('day.csv')
    
#     df_bike.temp = df_bike.temp.apply(lambda x: roundup_to(x,5))
#     df_bike.hum = df_bike.hum.apply(lambda x: roundup_to(x, 10))
#     df_bike.windspeed = df_bike.windspeed.apply(lambda x: roundup_to(x, 5))

#     df_bike.temp = df_bike.temp.apply(lambda x: roundup_to(x,5))
#     df_bike.hum = df_bike.hum.apply(lambda x: roundup_to(x, 10))
#     df_bike.windspeed = df_bike.windspeed.apply(lambda x: roundup_to(x, 5))
    
    # Map the numbers with actual values
    df_bike.season = df_bike.season.map({1:'spring', 2:'summer', 3:'fall', 4:'winter'})
    df_bike.yr = df_bike.yr.map({1:'2019', 0:'2018'})
    df_bike.weekday = df_bike.weekday.map({0:'sunday',1:'monday',2:'tuesday',3:'wednesday',4:'thursday',5:'friday', 6:'saturday'}) # Assumption 0 is Sunday, 1 is Monday and follows
    df_bike.weathersit = df_bike.weathersit.map({1:'cloud',2:'mist',3:'light',4:'heavy'})
    df_bike.mnth = df_bike.mnth.map({1:'jan',2:'feb',3:'mar',4:'apr',5:'may',6:'jun', 7:'jul', 8:'aug', 9:'sep', 10:'oct', 11:'nov', 12:'dec'})
    
    #Creating Dummies for categorical columns
    cat_col = ['season','yr','mnth','weathersit','weekday']
    dummy_col = pd.get_dummies(df_bike[cat_col], drop_first=True)
    df_bike = pd.concat([df_bike, dummy_col], axis = 1)
    
    #drop original columns
    df_bike = df_bike.drop(cat_col, axis = 1)
#     df_bike = df_bike.drop(['yr_2018'], axis =1)
    
    #drop columns which are not useful
    not_usefull_cols = ['instant','dteday','atemp','casual','registered']
    df_bike = df_bike.drop(not_usefull_cols, axis = 1)
    
    return df_bike
```

```python
df_bike = preprocessing()
```

### * statmodel_linear

This method will calculated the linear regression using statmodel
> - Prints the Summary of Linear Statmodel
> - Returns the linear statmodel 

```python
def statmodel_linear(X, y):
        # Add a constant
    X_sm = sm.add_constant(X)

    # Create a first fitted model
    linear_statmodel = sm.OLS(y, X_sm).fit()
    #Summary
    print(linear_statmodel.summary())
    
    return linear_statmodel
```

### * find_vif

This method will be used to find the VIF

```python
def find_vif(X):
    vif = pd.DataFrame()
    vif['Features'] = X.columns
    vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    print(vif)
    return vif
```

### * plot_model_stats
This function will be used to plot the following diagrams of the model
> - Error Distribution plot
> - Residual plot
> - High Coefficient features
> - VIF

```python
def plot_model_stats(y_train, y_train_pred, model_lr_sm, model_vif, title):
    fig = plt.figure(figsize=(14, 21))
    #Error Distribution plot
    plt.subplot(3,2,1)
    fig1 = sns.distplot((y_train - y_train_pred), bins = 20, color='b')
    fig1.title.set_text('Error Distriubtion Terms')                  # Plot heading 
    plt.xlabel('Errors', fontsize = 18)                         # X-label

    #y_train vs y_train_pred 
    plt.subplot(3,2,2)
    fig2 = sns.scatterplot(y_train,y_train_pred, color='g')
    fig2.title.set_text('y_train vs y_train_pred')              # Plot heading 
    plt.xlabel('y_train', fontsize=18)                          # X-label
    plt.ylabel('y_train_pred', fontsize=16)  

    #High Coefficient features
    plt.subplot(3,2,3)
    model_lr_sm.params = model_lr_sm.params.sort_values(ascending = False)
    model_lr_sm.params.drop(labels='const',inplace= True)
    fig2 = sns.barplot(model_lr_sm.params.index, model_lr_sm.params.values)
    fig2.title.set_text('Feature Coefficiency')              # Plot heading 
    plt.xlabel('Features', fontsize=18)                          # X-label
    plt.ylabel('Coefficiency', fontsize=16)  
    plt.xticks(rotation=90)
    fig.suptitle("Model6", fontsize=24, color ='r')

    #VIF
    plt.subplot(3,2,4)
    fig2 = sns.barplot(model_vif.Features, model_vif.VIF)
    fig2.title.set_text('VIF')              # Plot heading 
    plt.xlabel('Features', fontsize=18)                          # X-label
    plt.ylabel('VIF', fontsize=16)  
    plt.xticks(rotation=90)

    
    #Residual plot
    plt.subplot(3,2,5)
    fig2 = sns.scatterplot(y_train,(y_train - y_train_pred), color='g')
    fig2.title.set_text('Residuals Plot')              # Plot heading 
    plt.xlabel('y_train', fontsize=18)                          # X-label
    plt.ylabel('y_train - y_train_pred', fontsize=16)  
    
    plt.subplot(3,2,6)
    result = {'R-Squared' : round(model_lr_sm.rsquared,3), 'Adj. R-squared' : round(model_lr_sm.rsquared_adj,3)}
    plt.bar(*zip(*result.items()))
    plt.xlabel('Matrics', fontsize=18)                          # X-label
    plt.ylabel('Result', fontsize=16)
    plt.suptitle('Matrics Result')
    plt.show()

    fig.suptitle(title, fontsize=24, color ='r')
    plt.show()
```
